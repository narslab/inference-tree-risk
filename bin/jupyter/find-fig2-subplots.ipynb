```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMAGE_ARRAY = np.load('../../data/tidy/preprocessed-images/testing-w-252px-h-252px-scenario-Pr_Po_Im-train.npy',allow_pickle = True)
```

```python
def exploreProcessedImages(image_array):
    rows= 15
    cols = 15
    f, ax = plt.subplots(rows,cols, figsize=(50,30))
    k = 0 #1120 #1120 #2240
    for i in np.arange(rows):
        for j in np.arange(cols):
            ax[i][j].set_title(str(k))
            ax[i][j].imshow(image_array[k][0])#, cmap = 'gist_gray')
            ax[i][j].set_xticklabels([])
            ax[i][j].set_yticklabels([])
            k += 1
            plt.show()
    return
```

```python
exploreProcessedImages(IMAGE_ARRAY)
```
